### file

__换行__代表两个字符大小【windows系统下】

```python
f = open("./testfile/file",'r',encoding="utf-8")
# read后面读取的是5个字符。
data = f.read(10)
f.close()
print(data)
```

```python
f = open("./testfile/file",'w',encoding="utf-8")
# read后面读取的是5个字符。
f.write("hello world \n")
f.write("Tom Mary")

# 文件名
print(f.name)
print(f.fileno())
f.close()
```

- flush() 刷新缓存区
- fileno() 返回文件描述符
- isatty()  --不明白意思---用的比较少
- next()  --返回文件的下一行
- readline()  --读取整行，如果加上参数，和read()方法一致。
- readlines()  --读取所有行，返回列表
- seek()   --用于移动文件读取指针到指定位置
- tell()  --返回文件的当前位置，即指针当前位置

```python
f = open("./testfile/file",'r',encoding="utf-8")

for index in range(5):
    line = next(f)
    print("第%d行 - %s" % (index+1,line))
    
f.close()
```

```python
# readline()

f = open("./testfile/file",'r',encoding="utf-8")

# 回车好像是占用两个字符
for index in range(5):
    line = f.readline(2)
    print("第%d行 - %s" % (index+1,line))
    
f.close()
```

```python
# readlines()

f = open("./testfile/file",'r',encoding="utf-8")

# 回车好像是占用两个字符
# 注意这里的对列表添加索引index
# readlines是将所有内容读取出来放在内存，
for index,line in enumerate(f.readlines()):
    print("第%d行 - %s" % (index+1,line))
    
f.close()
```

```python
f = open("./testfile/file",'r',encoding="utf-8")

# f是文件对象，这里for内部将f做成一个迭代器，然后每读取一行的时候就去调用，相当于用时间换取空间
for i in f:
    print(i.strip())
    
f.close()
```

```python
# seek
# 第一个参数指定要偏移的--字节数--，第二个参数表示偏移的参考位置。
# 当传送比较大的文件的时候，中间断了，此时就比较有用
f = open("./testfile/file",'r',encoding="utf-8")
line = f.readline()
print(line)
f.seek(3,0)
print("当前位置",f.tell())
print(f.readline())
f.close()
```

```python
# tell()
# 中文占3个，注意和read()是按照字符来读，无论英文中文。
f = open("./testfile/file",'r',encoding="utf-8")
print(f.tell())
print(f.read(3))
print(f.tell())
```

```python
# truncate()  没有返回值，用的比较少
f = open("./testfile/file",'r+',encoding="utf-8")

# 带参数时，从文件开头截取到指定位置,后面的内容不要
# 截取是按照字节来截取的，因此可能出现错误，去掉encoding时，如果文件中存在汉字，那么会乱码
# f.truncate(20)
# line = f.read()
# print(line)


# 不带参数时,从指针当前位置开始删除，删除后面所有内容
f.seek(5)
f.truncate()
f.close()

```

```python
# writelines()
# 写一个序列
f = open("./testfile/file","r+",encoding="utf-8")
lists = ["菜鸟教程 \n","rooboo \n"]
f.writelines(lists)
f.close()
```

### JSON

如果我们要在不同的编程语言之间传递对象，就必须把对象序列化为标准格式，比如XML，但更好的方法是序列化为JSON，因为JSON表示出来就是一个字符串，可以被所有语言读取，也可以方便地存储到磁盘或者通过网络传输。JSON不仅是标准格式，并且比XML更快，而且可以直接在Web页面中读取，非常方便。

```python
import json
d = dict(name="bob",age=20,score=80)
# print(d)
json_str = json.dumps(d)  #返回一个str，内容是标准的sjon
json = json.loads(json_str)
print(json)
```

Python的dict对象可以直接序列化为JSON的{}，不过，很多时候，我们更喜欢用class表示对象，比如定义Student类，然后序列化：

```python
import json
class Student(object):
    def __init__(self,name,age,score):
        self.name = name;
        self.age = age
        self.score = score
s = Student('Tom',20,90)

#Student不是一个可以序列化的JSON对象，需要专门写一个转化函数
def student2dict(std):
    return{
        'name':std.name,
        'age':std.age,
        'score':std.score
    }
json_str = json.dumps(s,default=student2dict)
print(json_str)

# 转换成Student对象
def dict2student(d):
    return Student(d['name'],d['age'],d['score'])

print(json.loads(json_str,object_hook=dict2student))
```

### 文件读写

通常都是使用with语句来实现文件的打开操作，会自动调用close()方法

```python
# 读取UTF-8编码的文本文件时
with open('./testfile/foo.txt','r',encoding="UTF8")as f:
    print(f.read())

with open('./testfile/module_1.png','rb')as f:
    print(f.read())

# 遇到非法编码字符时，errors参数

with open('./testfile/foo.txt','r',encoding="gbk",errors='ignore')as f:
    print(f.read())
```
