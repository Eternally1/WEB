### python3输入输出

```python
s = "hello world"
s1 = repr(s)
print(s1)

hello = "hello world \n"
print(hello)
# repr可以转移字符串中特殊字符
hellos = repr(hello)
print(hellos)
```

```python
# rjust()让字符串靠右，并在左边填充空格  
# 还有ljust()  center()  zfill()
for x in range(1,11):
#     print(repr(x).rjust(2),repr(x*x).rjust(3),end=" ")
#     print(repr(x*x*x).rjust(4))
# 也可以使用上面这一种方法
      print(str(x).rjust(2),str(x*x).rjust(3),str(x*x*x).rjust(4))
```
